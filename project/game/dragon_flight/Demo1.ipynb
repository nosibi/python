```python
# outline
# 캐릭터는 좌우로만 피하고 적은 랜덤하게 나타나서 y=0에서부터 y = screen_height - enemy_height 까지 이동
# 닿지 않고 카운트 내로 생존하면 게임 승리


import pygame
pygame.init()
import random

# 스크린
screen_width = 1280
screen_height = 960
screen = pygame.display.set_mode((screen_width,screen_height))

title = pygame.display.set_caption("Demo1")

# Frame Per Second
clock = pygame.time.Clock()

# 배경
background = pygame.image.load(r'C:\python\project\game\dragon_flight\image_folder\background.png')

game_font = pygame.font.Font(None,100)
total_time = 20
start_ticks = pygame.time.get_ticks()


# 캐릭터
character = pygame.image.load(r'C:\python\project\game\dragon_flight\image_folder\character.png')
character_size = character.get_rect().size
character_width = character_size[0]
character_height = character_size[1]
character_xpos = (screen_width/2) - (character_width/2)
character_ypos = screen_height - (character_height * 2)

character_to_x = 0
character_speed = 10

# 적
enemy1 = pygame.image.load(r'C:\python\project\game\dragon_flight\image_folder\enemy1.png')
enemy1_size = enemy1.get_rect().size
enemy1_width = enemy1_size[0]
enemy1_height = enemy1_size[1]
enemy1_xpos = random.uniform(0,screen_width - enemy1_width)
enemy1_ypos = 0
enemy1_to_y = 0
enemy1_speed = 10

enemy2 = pygame.image.load(r'C:\python\project\game\dragon_flight\image_folder\enemy2.png')
enemy2_size = enemy2.get_rect().size
enemy2_width = enemy2_size[0]
enemy2_height = enemy2_size[1]
enemy2_xpos = random.uniform(0,screen_width - enemy2_width)
enemy2_ypos = 0
enemy2_to_y = 0
enemy2_speed = 10

enemy3 = pygame.image.load(r'C:\python\project\game\dragon_flight\image_folder\enemy3.png')
enemy3_size = enemy3.get_rect().size
enemy3_width = enemy3_size[0]
enemy3_height = enemy3_size[1]
enemy3_xpos = random.uniform(0,screen_width - enemy3_width)
enemy3_ypos = 0
enemy3_to_y = 0
enemy3_speed = 10

# 무기
weapon = pygame.image.load(r"C:\python\project\game\dragon_flight\image_folder\weapon.png")
weapon_size = weapon.get_rect().size
weapon_width = weapon_size[0]
weapon_height = weapon_size[1]
weapon_speed = 10
weapons = []


# run time
running = True
while running:
    dt = clock.tick(60)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_RIGHT:
                character_to_x += character_speed
            elif event.key == pygame.K_LEFT:
                character_to_x -= character_speed
            elif event.key == pygame.K_SPACE:
                weapon_xpos = character_xpos + (character_width / 2) - (weapon_width / 2)
                weapon_ypos = character_ypos
                weapons.append([weapon_xpos, weapon_ypos])
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                character_to_x = 0
                
    character_xpos += character_to_x
    
    
    # 무기 위치 변화
    weapons = [[w[0],(w[1] - weapon_speed)] for w in weapons]
    
    weapons = [[w[0],w[1]] for w in weapons if w[1] > 0]
    
    
    # 적 이동
    enemy1_to_y = 0.5 * dt
    enemy2_to_y = 0.5 * dt
    enemy3_to_y = 0.5 * dt
    
    enemy1_ypos += enemy1_to_y
    enemy2_ypos += enemy2_to_y
    enemy3_ypos += enemy3_to_y
    
    # 구역 제한
    if character_xpos < 0:
        character_xpos = 0
    elif character_xpos > screen_width - character_width:
        character_xpos = screen_width - character_width
        
        
    if enemy1_ypos > screen_height - enemy1_height:
        enemy1_ypos = 0
        enemy1_xpos = random.uniform(0,screen_width - enemy1_width)
    if enemy2_ypos > screen_height - enemy2_height:
        enemy2_ypos = 0
        enemy2_xpos = random.uniform(0,screen_width - enemy2_width)
    if enemy3_ypos > screen_height - enemy3_height:
        enemy3_ypos = 0
        enemy3_xpos = random.uniform(0,screen_width - enemy3_width)
        
        
    # 충돌처리
    character_rect = character.get_rect()
    character_rect.left = character_xpos
    character_rect.top = character_ypos
    enemy1_rect = enemy1.get_rect()
    enemy1_rect.left = enemy1_xpos
    enemy1_rect.top = enemy1_ypos
    enemy2_rect = enemy2.get_rect()
    enemy2_rect.left = enemy2_xpos
    enemy2_rect.top = enemy2_ypos
    enemy3_rect = enemy3.get_rect()
    enemy3_rect.left = enemy3_xpos
    enemy3_rect.top = enemy3_ypos
    
    
        
    if character_rect.colliderect(enemy1_rect):
        game_result = "Game Over"
        running = False
    elif character_rect.colliderect(enemy2_rect):
        game_result = "Game Over"
        running = False
    elif character_rect.colliderect(enemy3_rect):
        game_result = "Game Over"
        running = False
    if weapon_rect.colliderect(enemy1_rect):
        enemy1_ypos = 0
        enemy1_xpos = random.uniform(0,screen_width - enemy1_width)
    elif weapon_rect.colliderect(enemy2_rect):
        enemy2_ypos = 0
        enemy2_xpos = random.uniform(0,screen_width - enemy1_width)
    elif weapon_rect.colliderect(enemy3_rect):
        enemy3_ypos = 0
        enemy3_xpos = random.uniform(0,screen_width - enemy1_width)
        
    # 타이머
    elapsed_time = (pygame.time.get_ticks() - start_ticks) / 1000
    timer = game_font.render("Time : {}".format(int(total_time - elapsed_time)),True,(255,255,0))
    if total_time - elapsed_time <= 0:
        game_result = "Victory"
        running = False
        
    
    # 화면에 구현
    screen.blit(background,(0,0))
    for weapon_xpos, weapon_ypos in weapons:
        screen.blit(weapon,(weapon_xpos,weapon_ypos))
    screen.blit(character,(character_xpos,character_ypos))
    screen.blit(enemy1,(enemy1_xpos,enemy1_ypos))
    screen.blit(enemy2,(enemy2_xpos,enemy2_ypos))
    screen.blit(enemy3,(enemy3_xpos,enemy3_ypos))
    screen.blit(timer,(10,10))


    pygame.display.update()
    
msg = game_font.render(game_result,True,(255,255,255)) # 노란색
msg_rect = msg.get_rect(center = (screen_width / 2, screen_height / 2))
screen.blit(msg, msg_rect)
pygame.display.update()
    
    
pygame.time.delay(3000)
pygame.quit()
```
